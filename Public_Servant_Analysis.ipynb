### 使用的显著性检验方法

根据我以前学到过的：<br/>
* 海事局数据有两个自变量，一个因变量。两个自变量都是 multinominal varibles。此处应该选用Two-Way ANOVA（多自变量单因素方差分析）。
* 其他两个局，都是单个自变量和单个因变量，自变量是dummy or multinominal varibles，所以应该做T-test，以及One-Way ANOVA。
* 其实再深一步，也可以做完全多变量的分析，但因为很多自变量不是连续变量，因此没法用线性回归做拟合。

```python
from statsmodels.formula.api import ols

mod = ols(formula='竞争比 ~ 执法 + 专业类别', data=table_haishi_3)
res = mod.fit()
print(res.summary())
```

### 气象局 - 使用T-test
### 调查队 - 使用One-Way ANOVA

```python
table_qixiangju = table_analysis_1[table_analysis_1["用人司局"].str.contains("气象局")]
```

```python
table_qixiangju_a = table_qixiangju[["专业","职位简介","竞争比"]]
```

```python
list_s_2 = []
for item in table_qixiangju_a["专业"]:
    if "大气科学类" in item:
        list_s_2.append(1)
    else:
        list_s_2.append(0)
```

```python
table_qixiangju_a["是否是大气科学类"] = list_s_2
```

```python
import statsmodels.api as sm
x_ = table_qixiangju_a["是否是大气科学类"]
y_ = table_qixiangju_a["竞争比"]

res2 = sm.stats.ttest_ind(x_, y_)
print(res2)

### 此处显示为显著相关,使用t-test检验，因为自变量是个dummy variable，而因变量是个continuous variable
### p = 9.697220634126428e-13
```

#### 气象局的Pvalue是9.697220634126428e-1，小于 0.05
